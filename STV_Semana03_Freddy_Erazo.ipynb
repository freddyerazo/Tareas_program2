## Ejercicio 1. Herencia básica

Clase base `Animal` y clase hija `Perro`.

```python
# ============================================================
# EJERCICIO 1 - HERENCIA BÁSICA
# Se define la clase base 'Animal' con un constructor __init__
# que guarda el atributo 'nombre', y un método 'comer'.
# Luego se crea la clase hija 'Perro' que HEREDA de Animal.
# Perro tiene acceso a todos los métodos de Animal (herencia).
# ============================================================

class Animal:
    # Constructor: inicializa el atributo nombre
    def __init__(self, nombre):
        self.nombre = nombre  # Atributo de instancia

    # Método de la clase base
    def comer(self):
        print(f"{self.nombre} está comiendo.")


# Perro hereda de Animal (clase hija)
class Perro(Animal):
    # Método propio de Perro (no existe en Animal)
    def ladrar(self):
        print(f"{self.nombre} dice: guau guau")
```

```python
# ============================================================
# PRUEBA EJERCICIO 1
# Se crea un objeto Perro llamado 'Tobias'.
# .comer() → método HEREDADO de Animal
# .ladrar() → método PROPIO de Perro
# ============================================================
mi_perro = Perro("Tobias")   # Instanciación del objeto
mi_perro.comer()             # Llama método heredado de Animal
mi_perro.ladrar()            # Llama método propio de Perro
```

## Ejercicio 2. Herencia con atributos compartidos

```python
# ============================================================
# EJERCICIO 2 - HERENCIA CON ATRIBUTOS COMPARTIDOS
# Clase base 'Persona' con nombre y edad.
# Clase hija 'Estudiante' repite los atributos manualmente
# (sin usar super(), lo que se mejora en el Ejercicio 3).
# ============================================================

class Persona:
    # Constructor de la clase base
    def __init__(self, nombre, edad):
        self.nombre = nombre  # Atributo heredable
        self.edad = edad      # Atributo heredable

    # Método para mostrar datos de la persona
    def mostrar_datos(self):
        print(f"Nombre: {self.nombre} | Edad: {self.edad}")


class Estudiante(Persona):
    # Constructor de la clase hija (sin super() - forma manual)
    def __init__(self, nombre, edad, carrera):
        self.nombre = nombre   # Se repite el atributo de Persona
        self.edad = edad       # Se repite el atributo de Persona
        self.carrera = carrera # Atributo propio de Estudiante
```

```python
# ============================================================
# PRUEBA EJERCICIO 2
# Se crea un Estudiante y se usan métodos heredados y atributos propios.
# ============================================================
est = Estudiante("Freddy", 46, "Ciencia de Datos e Inteligencia Artificial")
est.mostrar_datos()           # Método heredado de Persona
print(f"Carrera: {est.carrera}")  # Atributo propio de Estudiante
```

## Ejercicio 3. Uso de `super()`

```python
# ============================================================
# EJERCICIO 3 - USO DE super()
# Mejora del Ejercicio 2: en lugar de repetir atributos,
# se usa super().__init__() para reutilizar el constructor
# de la clase padre Persona.
# ============================================================

class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def mostrar_datos(self):
        print(f"Nombre: {self.nombre} | Edad: {self.edad}")


class Estudiante(Persona):
    def __init__(self, nombre, edad, carrera):
        # super() llama al __init__ de Persona: evita repetir código
        super().__init__(nombre, edad) 
        self.carrera = carrera  # Atributo propio de Estudiante

    # Método adicional propio de Estudiante
    def mostrar_carrera(self):
        print(f"Carrera: {self.carrera}")
```

```python
# ============================================================
# PRUEBA EJERCICIO 3
# super() permite inicializar nombre y edad desde Persona
# ============================================================
est = Estudiante("Ana", 22, "Inteligencia Artificial")
est.mostrar_datos()     # Método heredado de Persona
est.mostrar_carrera()   # Método propio de Estudiante
```

## Ejercicio 4. Polimorfismo básico

```python
# ============================================================
# EJERCICIO 4 - POLIMORFISMO BÁSICO
# Varias clases hijas sobreescriben el mismo método 'hacer_sonido'.
# Cada clase tiene su propia implementación del método.
# Esto es polimorfismo: mismo nombre de método, distinto comportamiento.
# ============================================================

class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    # Método base que será sobrescrito por las clases hijas
    def hacer_sonido(self):
        print(f"{self.nombre} hace un sonido.")


# Clase Perro: sobreescribe hacer_sonido (polimorfismo)
class Perro(Animal):
    def hacer_sonido(self):
        print(f"{self.nombre} dice: ¡Guau!")


# Clase Gato: sobreescribe hacer_sonido (polimorfismo)
class Gato(Animal):
    def hacer_sonido(self):
        print(f"{self.nombre} dice: ¡Miau!")
```

```python
# ============================================================
# PRUEBA EJERCICIO 4
# Mismo método hacer_sonido(), diferente resultado según el objeto.
# ============================================================
perro = Perro("Rex")   # Objeto de tipo Perro
gato = Gato("Chipi")    # Objeto de tipo Gato
perro.hacer_sonido()   # Imprime sonido de perro
gato.hacer_sonido()    # Imprime sonido de gato
```

## Ejercicio 5. Polimorfismo con listas

```python
# ============================================================
# EJERCICIO 5 - POLIMORFISMO CON LISTAS
# Se guarda una lista con objetos de distintas clases.
# El bucle for llama al mismo método en cada objeto.
# Python ejecuta la versión correcta según el tipo real del objeto.
# ============================================================
animales = [
    Perro("Tobias"),   # Objeto Perro
    Gato("Chipi"),   # Objeto Gato
    Perro("Bobi"),   # Objeto Perro
    Gato("Michu"),   # Objeto Gato
]

# Polimorfismo: mismo método, comportamiento diferente según el objeto
for animal in animales:
    animal.hacer_sonido()  # Python elige la versión correcta automáticamente
```

## Ejercicio 6. Caso aplicado: Vehículos

```python
# ============================================================
# EJERCICIO 6 - CASO APLICADO: VEHÍCULOS
# Clase base Vehiculo con marca y modelo.
# Auto y Moto usan super() en __init__ y en mostrar_info().
# super().mostrar_info() reutiliza la impresión base y agrega detalles.
# ============================================================

class Vehiculo:
    # Constructor base
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    # Método base que imprime marca y modelo
    def mostrar_info(self):
        print(f"Vehículo: {self.marca} {self.modelo}")


class Auto(Vehiculo):
    def __init__(self, marca, modelo, puertas):
        super().__init__(marca, modelo)  # Hereda atributos base
        self.puertas = puertas           # Atributo propio

    def mostrar_info(self):
        super().mostrar_info()           # Reutiliza impresión base
        print(f"  Tipo: Auto | Puertas: {self.puertas}")


class Moto(Vehiculo):
    def __init__(self, marca, modelo, cilindrada):
        super().__init__(marca, modelo)  # Hereda atributos base
        self.cilindrada = cilindrada     # Atributo propio

    def mostrar_info(self):
        super().mostrar_info()           # Reutiliza impresión base
        print(f"  Tipo: Moto | Cilindrada: {self.cilindrada} cc")


# Prueba: se crean objetos y se llama mostrar_info()
auto = Auto("Volkwagen", "T Cross", 4)
moto = Moto("Honda", "500", 500)
auto.mostrar_info()   # Muestra info del auto
moto.mostrar_info()   # Muestra info de la moto
```

## Ejercicio 7. Sobrescritura de métodos

```python
# ============================================================
# EJERCICIO 7 - SOBRESCRITURA DE MÉTODOS
# Se agrega el método encender() a Auto y Moto.
# Cada subclase tiene su propia versión de encender().
# Esto es sobrescritura (override): mismo nombre, diferente cuerpo.
# ============================================================

class Auto(Vehiculo):
    def __init__(self, marca, modelo, puertas):
        super().__init__(marca, modelo)  # Constructor heredado
        self.puertas = puertas

    def mostrar_info(self):
        super().mostrar_info()
        print(f"  Tipo: Auto | Puertas: {self.puertas}")

    # Método sobrescrito: comportamiento específico para Auto
    def encender(self):
        print(f"El auto {self.marca} {self.modelo} enciende con llave electrónica.")


class Moto(Vehiculo):
    def __init__(self, marca, modelo, cilindrada):
        super().__init__(marca, modelo)  # Constructor heredado
        self.cilindrada = cilindrada

    def mostrar_info(self):
        super().mostrar_info()
        print(f"  Tipo: Moto | Cilindrada: {self.cilindrada} cc")

    # Método sobrescrito: comportamiento específico para Moto
    def encender(self):
        print(f"La moto {self.marca} {self.modelo} enciende con patada de arranque.")
```

```python
# ============================================================
# PRUEBA EJERCICIO 7
# Mismo método encender(), diferente resultado (sobrescritura).
# ============================================================
auto = Auto("Toyota", "Corolla", 4)
moto = Moto("Honda", "CB500", 500)
auto.encender()   # Versión de Auto
moto.encender()   # Versión de Moto
```

## Ejercicio 8. `super()` dentro de un método

```python
# ============================================================
# EJERCICIO 8 - super() DENTRO DE UN MÉTODO
# Docente hereda de Empleado.
# En presentarse(), primero llama super().presentarse() para
# reutilizar el mensaje base y luego agrega su propio mensaje.
# ============================================================

class Empleado:
    def __init__(self, nombre, sueldo):
        self.nombre = nombre
        self.sueldo = sueldo

    # Método base de presentación
    def presentarse(self):
        print(f"Hola, soy {self.nombre} y mi sueldo es ${self.sueldo:.2f}")


class Docente(Empleado):
    def __init__(self, nombre, sueldo, materia):
        super().__init__(nombre, sueldo)  # Reutiliza constructor base
        self.materia = materia

    def presentarse(self):
        super().presentarse()  # Llama presentación base de Empleado
        print(f"Soy docente y enseño {self.materia}.")
```

```python
# ============================================================
# PRUEBA EJERCICIO 8
# presentarse() combina el mensaje del padre y el propio.
# ============================================================
docente = Docente("Ing. Zuñiga", 2500, "Programación II")
docente.presentarse()  # Ejecuta super().presentarse() + mensaje propio
```

## Ejercicio 9. Herencia multinivel

```python
# ============================================================
# EJERCICIO 9 - HERENCIA MULTINIVEL
# Cadena: Persona → Estudiante → EstudianteBecado
# Cada nivel usa super().__init__() para pasar atributos hacia arriba.
# EstudianteBecado accede a atributos de Persona, Estudiante y propios.
# ============================================================

# Nivel 1: Clase base
class Persona:
    def __init__(self, nombre, cedula):
        self.nombre = nombre
        self.cedula = cedula


# Nivel 2: Hereda de Persona
class Estudiante(Persona):
    def __init__(self, nombre, cedula, carrera, nivel):
        super().__init__(nombre, cedula)  # Sube a Persona
        self.carrera = carrera
        self.nivel = nivel


# Nivel 3: Hereda de Estudiante (herencia multinivel)
class EstudianteBecado(Estudiante):
    def __init__(self, nombre, cedula, carrera, nivel, tipo_beca, porcentaje_descuento):
        super().__init__(nombre, cedula, carrera, nivel)  # Sube a Estudiante
        self.tipo_beca = tipo_beca
        self.porcentaje_descuento = porcentaje_descuento

    # Muestra todos los atributos acumulados en la cadena de herencia
    def mostrar_info(self):
        print(f"Nombre    : {self.nombre}")              # De Persona
        print(f"Cédula    : {self.cedula}")              # De Persona
        print(f"Carrera   : {self.carrera}")             # De Estudiante
        print(f"Nivel     : {self.nivel}")               # De Estudiante
        print(f"Beca      : {self.tipo_beca}")           # Propio
        print(f"Descuento : {self.porcentaje_descuento}%")  # Propio
```

```python
# ============================================================
# PRUEBA EJERCICIO 9
# Se crea un EstudianteBecado con datos de los 3 niveles.
# ============================================================
becado = EstudianteBecado(
    "Carlos López", "1712345678",    # Atributos de Persona
    "Ciencia de Datos", "Segundo",   # Atributos de Estudiante
    "Excelencia Académica", 50        # Atributos propios
)
becado.mostrar_info()  # Muestra toda la información acumulada
```

## Ejercicio 10. Reto final

```python
# ============================================================
# EJERCICIO 10 - RETO FINAL
# Sistema completo con herencia, super() y polimorfismo.
# Clase base: Empleado (nombre, sueldo, mostrar_datos, calcular_bono)
# Subclases: Docente y Administrativo con bono diferente cada uno.
# Se usa una lista para aplicar polimorfismo en el reporte.
# ============================================================

# Clase base con método calcular_bono() retornando 0 por defecto
class Empleado:
    def __init__(self, nombre, sueldo):
        self.nombre = nombre
        self.sueldo = sueldo

    def mostrar_datos(self):
        print(f"Nombre : {self.nombre} | Sueldo: ${self.sueldo:.2f}")

    # Método base del bono: devuelve 0 (se sobreescribirá)
    def calcular_bono(self):
        return 0


# Docente: bono del 15% del sueldo
class Docente(Empleado):
    def __init__(self, nombre, sueldo, materia):
        super().__init__(nombre, sueldo)  # Hereda constructor
        self.materia = materia

    def mostrar_datos(self):
        super().mostrar_datos()           # Reutiliza impresión base
        print(f"  Rol: Docente | Materia: {self.materia}")

    def calcular_bono(self):
        return self.sueldo * 0.15         # Bono: 15%


# Administrativo: bono del 10% del sueldo
class Administrativo(Empleado):
    def __init__(self, nombre, sueldo, departamento):
        super().__init__(nombre, sueldo)  # Hereda constructor
        self.departamento = departamento

    def mostrar_datos(self):
        super().mostrar_datos()           # Reutiliza impresión base
        print(f"  Rol: Administrativo | Depto: {self.departamento}")

    def calcular_bono(self):
        return self.sueldo * 0.10         # Bono: 10%


# Lista de empleados - polimorfismo: objetos distintos, mismo bucle
empleados = [
    Docente("Ing. Zuñiga", 2500, "Programación II"),
    Docente("Dra. Torres", 3000, "Estadística"),
    Administrativo("Srta. Pérez", 1800, "Registro Académico"),
    Administrativo("Sr. López", 2000, "Financiero"),
]

# Recorre la lista aplicando polimorfismo en mostrar_datos() y calcular_bono()
print("===== Reporte de Empleados =====")
for emp in empleados:
    emp.mostrar_datos()                      # Polimorfismo: versión según tipo
    print(f"  Bono: ${emp.calcular_bono():.2f}")  # Polimorfismo: bono según tipo
    print("-" * 40)
```
