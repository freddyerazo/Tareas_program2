## Ejercicio 1. Clase `Estudiante`

Crea una clase llamada `Estudiante` con los siguientes atributos:

- nombre
- nota1
- nota2

Métodos requeridos:

- `calcular_promedio()`
- `aprobo()`: debe indicar si el estudiante aprobó o no
- `mostrar_info()`

### Reglas
- El promedio es la media de las dos notas
- Si el promedio es mayor o igual a 7, aprueba

```python
class Estudiante:
    def __init__(self, nombre, nota1, nota2):
        self.nombre = nombre
        self.nota1 = nota1
        self.nota2 = nota2

    ## Retorna el promedio de las dos notas.
    def calcular_promedio(self):
        return (self.nota1 + self.nota2) / 2

    ## Retorna True si el promedio es mayor o igual a 7.
    def aprobo(self):
        if self.calcular_promedio() >= 7:
            return True
        return False

    ## Muestra la información completa del estudiante.
    def mostrar_info(self):
        promedio = self.calcular_promedio()
        estado = "APROBADO" if self.aprobo() else "REPROBADO"
        print(f"Estudiante : {self.nombre}")
        print(f"Nota 1     : {self.nota1}")
        print(f"Nota 2     : {self.nota2}")
        print(f"Promedio   : {promedio:.2f}")
        print(f"Estado     : {estado}")


# Impresion en Pantalla de informacion y calculos
e1 = Estudiante("Ana Lopez", 8, 9)
e1.mostrar_info()
print()
e2 = Estudiante("Carlos Ruiz", 5, 6)
e2.mostrar_info()
print()
e3 = Estudiante("Freddy Bueno", 3, 2)
e3.mostrar_info()

```

## Ejercicio 2. Clase `Producto`

Crea una clase llamada `Producto` con:

- nombre
- precio
- stock

Métodos:

- `aplicar_descuento(porcentaje)`
- `vender(cantidad)`
- `mostrar_producto()`

### Reglas
- El descuento no puede ser negativo ni mayor a 100
- No se puede vender más unidades de las disponibles en stock
- El precio nunca debe quedar negativo

```python
class Producto:
    def __init__(self, nombre, precio, stock):
        self.nombre = nombre
        self.precio = precio
        self.stock = stock

    def aplicar_descuento(self, porcentaje):
        if porcentaje < 0 or porcentaje > 100:
            print("Error: el porcentaje de descuento debe estar entre 0 y 100.")
            return
        descuento = self.precio * (porcentaje / 100)
        self.precio = self.precio - descuento
        if self.precio < 0:
            self.precio = 0
        print(f"Descuento del {porcentaje}% aplicado. Nuevo precio: ${self.precio:.2f}")

    def vender(self, cantidad):
        if cantidad <= 0:
            print("Error: la cantidad a vender debe ser mayor a 0.")
            return
        if cantidad > self.stock:
            print(f"Error: no hay suficiente stock. Disponible: {self.stock} unidades.")
            return
        self.stock -= cantidad
        print(f"Venta exitosa: {cantidad} unidad(es) de '{self.nombre}'. Stock restante: {self.stock}")

    def mostrar_producto(self):
        print(f"Producto : {self.nombre}")
        print(f"Precio   : ${self.precio:.2f}")
        print(f"Stock    : {self.stock} unidades")


# Impresion de resultados

p = Producto("Iphone", 1600, 6)

print("--- Estado inicial ---")
p.mostrar_producto()
print()
p.aplicar_descuento(15)   # descuento válido
# p.aplicar_descuento(99)  # descuento inválido
print()
# p.vender(3)               # venta válida
p.vender(20)              # supera el stock
print()
# print("--- Estado final ---")
# p.mostrar_producto()
```

## Ejercicio 3. Clase `CuentaBancaria`

Diseña una clase `CuentaBancaria` con:

- titular
- saldo

Métodos:

- `depositar(monto)`
- `retirar(monto)`
- `transferir(otra_cuenta, monto)`
- `mostrar_info()`

### Reglas
- No se puede depositar ni retirar montos negativos
- No se puede retirar más dinero del saldo disponible
- La transferencia debe descontar de una cuenta y sumar a la otra

```python
class CuentaBancaria:
    def __init__(self, titular, saldo=0):
        self.titular = titular
        self.saldo = saldo

    ## Agrega un monto al saldo de la cuenta.
    def depositar(self, monto):
        if monto <= 0:
            print("Error: el monto a depositar debe ser mayor a 0.")
            return
        self.saldo += monto
        print(f"Depósito de ${monto:.2f} realizado. Saldo actual: ${self.saldo:.2f}")

    ## Extrae un monto del saldo si hay fondos suficientes.
    def retirar(self, monto):
        if monto <= 0:
            print("Error: el monto a retirar debe ser mayor a 0.")
            return
        if monto > self.saldo:
            print(f"Error: saldo insuficiente. Saldo disponible: ${self.saldo:.2f}")
            return
        self.saldo -= monto
        print(f"Retiro de ${monto:.2f} realizado. Saldo actual: ${self.saldo:.2f}")

    ## Transfiere un monto de esta cuenta a otra cuenta bancaria.
    def transferir(self, otra_cuenta, monto):
        if monto <= 0:
            print("Error: el monto a transferir debe ser mayor a 0.")
            return
        if monto > self.saldo:
            print(f"Error: saldo insuficiente para transferir. Saldo disponible: ${self.saldo:.2f}")
            return
        self.saldo -= monto
        otra_cuenta.saldo += monto
        print(f"Transferencia de ${monto:.2f} de '{self.titular}' a '{otra_cuenta.titular}' realizada.")

    ## Muestra el estado actual de la cuenta.
    def mostrar_info(self):
        print(f"Titular : {self.titular}")
        print(f"Saldo   : ${self.saldo:.2f}")


# --- Prueba ---
cuenta1 = CuentaBancaria("María García", 500)
cuenta2 = CuentaBancaria("Pedro Pérez", 200)

cuenta1.mostrar_info()
print()
cuenta1.depositar(300)       # depósito válido
cuenta1.retirar(100)         # retiro válido
cuenta1.retirar(2000)        # saldo insuficiente
print()
cuenta1.transferir(cuenta2, 250)  # transferencia válida
print()
print("--- Estado final ---")
cuenta1.mostrar_info()
print()
cuenta2.mostrar_info()
```

## Ejercicio 4. Clase `Asignatura`

Crea una clase `Asignatura` con:

- nombre
- paralelo
- estudiantes (lista)

Métodos:

- `agregar_estudiante(nombre_estudiante)`
- `eliminar_estudiante(nombre_estudiante)`
- `mostrar_estudiantes()`
- `total_estudiantes()`

### Objetivo
Trabajar con listas dentro de un objeto.


```python
class Asignatura:
    def __init__(self, nombre, paralelo):
        self.nombre = nombre
        self.paralelo = paralelo
        self.estudiantes = []

    ## Agrega un estudiante a la lista si no está ya inscrito.
    def agregar_estudiante(self, nombre_estudiante):
        if nombre_estudiante in self.estudiantes:
            print(f"'{nombre_estudiante}' ya está inscrito en la asignatura.")
        else:
            self.estudiantes.append(nombre_estudiante)
            print(f"'{nombre_estudiante}' agregado correctamente.")

    ## Elimina un estudiante de la lista si existe.
    def eliminar_estudiante(self, nombre_estudiante):
        if nombre_estudiante in self.estudiantes:
            self.estudiantes.remove(nombre_estudiante)
            print(f"'{nombre_estudiante}' eliminado de la asignatura.")
        else:
            print(f"'{nombre_estudiante}' no está inscrito en la asignatura.")

    ## Lista todos los estudiantes inscritos en la asignatura.
    def mostrar_estudiantes(self):
        print(f"Asignatura : {self.nombre} | Paralelo: {self.paralelo}")
        if not self.estudiantes:
            print("  (Sin estudiantes inscritos)")
        else:
            for i, nombre in enumerate(self.estudiantes, start=1):
                print(f"  {i}. {nombre}")

    ## Retorna el número total de estudiantes inscritos.
    def total_estudiantes(self):
        total = len(self.estudiantes)
        print(f"Total de estudiantes: {total}")
        return total


# --- Prueba ---
asig = Asignatura("Programación 2", "A")
asig.agregar_estudiante("Luisa Torres")
asig.agregar_estudiante("Miguel Mora")
asig.agregar_estudiante("Luisa Torres")  # duplicado
asig.agregar_estudiante("Sofía Vega")
print()
asig.mostrar_estudiantes()
print()
asig.eliminar_estudiante("Miguel Mora")
asig.eliminar_estudiante("Juan Nadie")   # no existe
print()
asig.mostrar_estudiantes()
asig.total_estudiantes()
```

## Ejercicio 5. Clase `Vehiculo`

Crea una clase `Vehiculo` con:

- marca
- modelo
- encendido (inicia en `False`)
- velocidad (inicia en `0`)

Métodos:

- `encender()`
- `apagar()`
- `acelerar(cantidad)`
- `frenar(cantidad)`
- `mostrar_estado()`

### Reglas
- No se puede acelerar si el vehículo está apagado
- La velocidad no puede ser negativa
- No se puede apagar el vehículo si aún tiene velocidad mayor que cero

```python
class Vehiculo:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo
        self.encendido = False
        self.velocidad = 0

    ## Enciende el vehículo si está apagado.
    def encender(self):
        if self.encendido:
            print("El vehículo ya está encendido.")
        else:
            self.encendido = True
            print(f"{self.marca} {self.modelo} encendido. ¡Listo para rodar!")

    ## Apaga el vehículo solo si la velocidad es 0.
    def apagar(self):
        if not self.encendido:
            print("El vehículo ya está apagado.")
        elif self.velocidad > 0:
            print(f"No se puede apagar: el vehículo va a {self.velocidad} km/h. Frena primero.")
        else:
            self.encendido = False
            print(f"{self.marca} {self.modelo} apagado.")

    ## Aumenta la velocidad solo si el vehículo está encendido.
    def acelerar(self, cantidad):
        if not self.encendido:
            print("No se puede acelerar: el vehículo está apagado.")
            return
        if cantidad <= 0:
            print("La cantidad a acelerar debe ser mayor a 0.")
            return
        self.velocidad += cantidad
        print(f"Acelerando... Velocidad actual: {self.velocidad} km/h")

    ## Reduce la velocidad sin bajar de 0.
    def frenar(self, cantidad):
        if cantidad <= 0:
            print("La cantidad a frenar debe ser mayor a 0.")
            return
        self.velocidad -= cantidad
        if self.velocidad < 0:
            self.velocidad = 0
        print(f"Frenando... Velocidad actual: {self.velocidad} km/h")

    ## Muestra el estado actual del vehículo.
    def mostrar_estado(self):
        estado = "ENCENDIDO" if self.encendido else "APAGADO"
        print(f"Vehículo  : {self.marca} {self.modelo}")
        print(f"Estado    : {estado}")
        print(f"Velocidad : {self.velocidad} km/h")


# --- Prueba ---
auto = Vehiculo("Toyota", "Corolla")
auto.mostrar_estado()
print()
auto.acelerar(50)        # no se puede, está apagado
auto.encender()
auto.acelerar(60)
auto.acelerar(40)
auto.apagar()            # no se puede, tiene velocidad
auto.frenar(80)          # frena hasta llegar a 0
auto.mostrar_estado()
print()
auto.apagar()            # ahora sí
```

## Ejercicio 6. Reto integrador

Desarrolla una clase llamada `Curso` con los siguientes atributos:

- nombre
- docente
- cupo_maximo
- inscritos (lista)

Métodos:

- `inscribir_estudiante(nombre)`
- `retirar_estudiante(nombre)`
- `hay_cupo()`
- `mostrar_resumen()`

### Reglas
- No se debe permitir inscribir más estudiantes que el cupo máximo
- No se debe repetir el nombre de un estudiante
- Mostrar cuántos cupos quedan disponibles

```python
class Curso:
    def __init__(self, nombre, docente, cupo_maximo):
        self.nombre = nombre
        self.docente = docente
        self.cupo_maximo = cupo_maximo
        self.inscritos = []

    ## Retorna True si aún hay cupos disponibles.
    def hay_cupo(self):
        return len(self.inscritos) < self.cupo_maximo

    ## Inscribe un estudiante si hay cupo y no está ya registrado.
    def inscribir_estudiante(self, nombre):
        if nombre in self.inscritos:
            print(f"'{nombre}' ya está inscrito en el curso.")
        elif not self.hay_cupo():
            print(f"No hay cupo disponible. El curso '{self.nombre}' está lleno.")
        else:
            self.inscritos.append(nombre)
            cupos_restantes = self.cupo_maximo - len(self.inscritos)
            print(f"'{nombre}' inscrito correctamente. Cupos restantes: {cupos_restantes}")

    ## Retira un estudiante del curso si está inscrito.
    def retirar_estudiante(self, nombre):
        if nombre in self.inscritos:
            self.inscritos.remove(nombre)
            print(f"'{nombre}' retirado del curso.")
        else:
            print(f"'{nombre}' no está inscrito en el curso.")

    ## Muestra un resumen del curso con inscritos y cupos disponibles.
    def mostrar_resumen(self):
        cupos_disponibles = self.cupo_maximo - len(self.inscritos)
        print(f"Curso          : {self.nombre}")
        print(f"Docente        : {self.docente}")
        print(f"Cupo máximo    : {self.cupo_maximo}")
        print(f"Inscritos      : {len(self.inscritos)}")
        print(f"Cupos libres   : {cupos_disponibles}")
        if self.inscritos:
            print("Lista de inscritos:")
            for i, nombre in enumerate(self.inscritos, start=1):
                print(f"  {i}. {nombre}")
        else:
            print("  (Sin estudiantes inscritos)")


# --- Prueba ---
curso = Curso("Programación 2", "Ing. Gómez", cupo_maximo=3)

curso.inscribir_estudiante("Ana")
curso.inscribir_estudiante("Luis")
curso.inscribir_estudiante("Ana")   # duplicado
curso.inscribir_estudiante("Sofía")
curso.inscribir_estudiante("Pablo") # cupo lleno
print()
curso.mostrar_resumen()
print()
curso.retirar_estudiante("Luis")
curso.inscribir_estudiante("Pablo") # ahora sí hay cupo
print()
curso.mostrar_resumen()
```
