## Ejercicio 1. Sistema de empleados universitarios

```python
# Clase base
class Empleado:
    def __init__(self, nombre, sueldo):
        self.nombre = nombre
        self.sueldo = sueldo

    def mostrar_datos(self):
        print(f"Nombre : {self.nombre}")
        print(f"Sueldo : ${self.sueldo:.2f}")

    def calcular_bono(self):
        return 0


# Clase hija: Docente
class Docente(Empleado):
    def __init__(self, nombre, sueldo, horas_clase):
        super().__init__(nombre, sueldo)
        self.horas_clase = horas_clase

    def mostrar_datos(self):
        super().mostrar_datos()
        print(f"Rol    : Docente | Horas clase: {self.horas_clase}")

    def calcular_bono(self):
        # Bono: 15% del sueldo + $5 por hora de clase
        return self.sueldo * 0.15 + self.horas_clase * 5


# Clase hija: Administrativo
class Administrativo(Empleado):
    def __init__(self, nombre, sueldo, departamento):
        super().__init__(nombre, sueldo)
        self.departamento = departamento

    def mostrar_datos(self):
        super().mostrar_datos()
        print(f"Rol    : Administrativo | Depto: {self.departamento}")

    def calcular_bono(self):
        # Bono: 10% del sueldo
        return self.sueldo * 0.10


# Lista de empleados y polimorfismo
empleados = [
    Docente("Ing. Zuñiga", 2500, 20),
    Docente("Dra. Torres", 3000, 16),
    Administrativo("Srta. Pérez", 1800, "Registro Académico"),
    Administrativo("Sr. López", 2000, "Financiero"),
]

print("===== Reporte de Empleados Universitarios =====")
for emp in empleados:
    emp.mostrar_datos()
    print(f"Bono   : ${emp.calcular_bono():.2f}")
    print("-" * 45)
```

## Ejercicio 2. Sistema de figuras con cálculo de área

```python
import math

class Figura:
    def calcular_area(self):
        return 0

    def describir(self):
        print("Figura geométrica")


class Rectangulo(Figura):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def calcular_area(self):
        return round(self.base * self.altura, 2)

    def describir(self):
        print(f"Rectángulo | Base: {self.base} | Altura: {self.altura} | Área: {self.calcular_area()}")


class Circulo(Figura):
    def __init__(self, radio):
        self.radio = radio

    def calcular_area(self):
        return round(math.pi * self.radio ** 2, 2)

    def describir(self):
        print(f"Círculo    | Radio: {self.radio} | Área: {self.calcular_area()}")


class Triangulo(Figura):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def calcular_area(self):
        return round(self.base * self.altura / 2, 2)

    def describir(self):
        print(f"Triángulo  | Base: {self.base} | Altura: {self.altura} | Área: {self.calcular_area()}")


# Lista de figuras con polimorfismo
figuras = [
    Rectangulo(5, 3),
    Circulo(4),
    Triangulo(6, 8),
]

print("===== Reporte de Figuras =====")
for figura in figuras:
    figura.describir()
```

## Ejercicio 3. Gestión de cuentas bancarias

```python
class CuentaBancaria:
    def __init__(self, titular, saldo):
        self.titular = titular
        self.saldo = saldo

    def depositar(self, monto):
        if monto <= 0:
            print("Error: el monto debe ser positivo.")
        else:
            self.saldo += monto
            print(f"Depósito de ${monto:.2f} exitoso. Saldo: ${self.saldo:.2f}")

    def retirar(self, monto):
        print("Método base: sin reglas definidas.")

    def mostrar_info(self):
        print(f"Titular: {self.titular} | Saldo: ${self.saldo:.2f}")


class CuentaAhorros(CuentaBancaria):
    def retirar(self, monto):
        if monto <= 0:
            print("Error: el monto debe ser positivo.")
        elif monto > self.saldo:
            print(f"Error: saldo insuficiente. Saldo disponible: ${self.saldo:.2f}")
        else:
            self.saldo -= monto
            print(f"Retiro de ${monto:.2f} exitoso. Saldo: ${self.saldo:.2f}")


class CuentaCorriente(CuentaBancaria):
    def __init__(self, titular, saldo, limite_sobregiro):
        super().__init__(titular, saldo)
        self.limite_sobregiro = limite_sobregiro

    def retirar(self, monto):
        if monto <= 0:
            print("Error: el monto debe ser positivo.")
        elif monto > self.saldo + self.limite_sobregiro:
            print(f"Error: supera el límite de sobregiro (${self.limite_sobregiro:.2f}).")
        else:
            self.saldo -= monto
            print(f"Retiro de ${monto:.2f} exitoso. Saldo: ${self.saldo:.2f}")


# Pruebas
print("=== Cuenta de Ahorros ===")
ca = CuentaAhorros("Freddy Erazo", 500)
ca.mostrar_info()
ca.depositar(200)
ca.retirar(100)
ca.retirar(800)  # Error esperado

print()
print("=== Cuenta Corriente ===")
cc = CuentaCorriente("Diana Cachago", 300, 500)
cc.mostrar_info()
cc.retirar(700)   # Usa sobregiro
cc.retirar(200)   # Supera límite - error esperado
```

## Ejercicio 4. Inventario de dispositivos tecnológicos

```python
class Dispositivo:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    def encender(self):
        print(f"{self.marca} {self.modelo} encendiendo...")

    def mostrar_info(self):
        print(f"Dispositivo: {self.marca} {self.modelo}")

    def tipo_uso(self):
        print("Uso general")


class Laptop(Dispositivo):
    def __init__(self, marca, modelo, ram_gb):
        super().__init__(marca, modelo)
        self.ram_gb = ram_gb

    def encender(self):
        print(f"Laptop {self.marca} {self.modelo} iniciando sistema operativo... ({self.ram_gb}GB RAM)")

    def tipo_uso(self):
        print(f"{self.marca} {self.modelo}: ideal para trabajo y desarrollo.")


class Telefono(Dispositivo):
    def __init__(self, marca, modelo, numero):
        super().__init__(marca, modelo)
        self.numero = numero

    def encender(self):
        print(f"Teléfono {self.marca} {self.modelo} activando línea {self.numero}...")

    def tipo_uso(self):
        print(f"{self.marca} {self.modelo}: comunicación y apps móviles.")


class Tablet(Dispositivo):
    def __init__(self, marca, modelo, pulgadas):
        super().__init__(marca, modelo)
        self.pulgadas = pulgadas

    def encender(self):
        print(f"Tablet {self.marca} {self.modelo} ({self.pulgadas}\") lista para usar.")

    def tipo_uso(self):
        print(f"{self.marca} {self.modelo}: entretenimiento y educación.")


# Lista de dispositivos - polimorfismo
inventario = [
    Laptop("Dell", "XPS 15", 16),
    Telefono("Samsung", "Galaxy S24", "0986686983"),
    Tablet("Apple", "iPad Air", 11),
]

print("===== Inventario de Dispositivos =====")
for dispositivo in inventario:
    dispositivo.encender()
    dispositivo.tipo_uso()
    print("-" * 45)
```

## Ejercicio 5. Plataforma académica con herencia multinivel

```python
class Persona:
    def __init__(self, nombre, cedula):
        self.nombre = nombre
        self.cedula = cedula


class Estudiante(Persona):
    def __init__(self, nombre, cedula, carrera, nivel):
        super().__init__(nombre, cedula)
        self.carrera = carrera
        self.nivel = nivel


class EstudianteBecado(Estudiante):
    def __init__(self, nombre, cedula, carrera, nivel, tipo_beca, porcentaje_descuento):
        super().__init__(nombre, cedula, carrera, nivel)
        self.tipo_beca = tipo_beca
        self.porcentaje_descuento = porcentaje_descuento

    def mostrar_info(self):
        print("===== Estudiante Becado =====")
        print(f"Nombre     : {self.nombre}")
        print(f"Cédula     : {self.cedula}")
        print(f"Carrera    : {self.carrera}")
        print(f"Nivel      : {self.nivel}")
        print(f"Beca       : {self.tipo_beca}")
        print(f"Descuento  : {self.porcentaje_descuento}%")


# Prueba
becado = EstudianteBecado(
    "Ariana Erazo", "1712345678",
    "Ciencia de Datos", "Segundo",
    "Excelencia Académica", 75
)
becado.mostrar_info()
```

## Ejercicio 6. Polimorfismo con reporte general

```python
class UsuarioSistema:
    def __init__(self, nombre):
        self.nombre = nombre

    def generar_reporte(self):
        print(f"Usuario: {self.nombre}")


class Estudiante(UsuarioSistema):
    def __init__(self, nombre, carrera, promedio):
        super().__init__(nombre)
        self.carrera = carrera
        self.promedio = promedio

    def generar_reporte(self):
        print(f"[ESTUDIANTE] {self.nombre} | Carrera: {self.carrera} | Promedio: {self.promedio}")


class Docente(UsuarioSistema):
    def __init__(self, nombre, materia, horas):
        super().__init__(nombre)
        self.materia = materia
        self.horas = horas

    def generar_reporte(self):
        print(f"[DOCENTE] {self.nombre} | Materia: {self.materia} | Horas: {self.horas}")


class Coordinador(UsuarioSistema):
    def __init__(self, nombre, facultad):
        super().__init__(nombre)
        self.facultad = facultad

    def generar_reporte(self):
        print(f"[COORDINADOR] {self.nombre} | Facultad: {self.facultad}")


# Lista de usuarios
usuarios = [
    Estudiante("Carlos López", "Ciencia de Datos", 9.2),
    Docente("Ing. Zuñiga", "Programación II", 20),
    Coordinador("Dr. Ramos", "Ingeniería"),
    Estudiante("Ana Torres", "IA", 8.7),
]

# Reto adicional: ordenar por nombre antes de imprimir
usuarios_ordenados = sorted(usuarios, key=lambda u: u.nombre)

print("===== Reporte General del Sistema =====")
for usuario in usuarios_ordenados:
    usuario.generar_reporte()
```

## Ejercicio 7. Uso de `super()` en métodos no constructores

```python
class Persona:
    def __init__(self, nombre, cedula):
        self.nombre = nombre
        self.cedula = cedula

    def presentarse(self):
        print(f"Hola, soy {self.nombre} | Cédula: {self.cedula}")


class Docente(Persona):
    def __init__(self, nombre, cedula, materia):
        super().__init__(nombre, cedula)
        self.materia = materia

    def presentarse(self):
        super().presentarse()  # reutiliza la presentación base
        print(f"Soy docente y enseño {self.materia}.")


class Investigador(Persona):
    def __init__(self, nombre, cedula, linea_investigacion):
        super().__init__(nombre, cedula)
        self.linea_investigacion = linea_investigacion

    def presentarse(self):
        super().presentarse()  # reutiliza la presentación base
        print(f"Soy investigador en el área de {self.linea_investigacion}.")


# Prueba
doc = Docente("Ing. Zuñiga", "1701234567", "Programación II")
inv = Investigador("Dra. Torres", "1709876543", "Machine Learning")

doc.presentarse()
print()
inv.presentarse()
```

## Ejercicio 8. Reto integrador

**Contexto elegido:** Sistema de logística de transporte (aplicado a Bellaflor Group)

```python
# Reto integrador: Sistema de transporte logístico - Bellaflor Group

class Transporte:
    def __init__(self, placa, capacidad_kg):
        self.placa = placa
        self.capacidad_kg = capacidad_kg

    def mostrar_info(self):
        print(f"Placa: {self.placa} | Capacidad: {self.capacidad_kg} kg")

    def calcular_costo(self, km):
        return 0

    def tipo_carga(self):
        print("Carga general")


class Furgon(Transporte):
    def __init__(self, placa, capacidad_kg, refrigerado):
        super().__init__(placa, capacidad_kg)
        self.refrigerado = refrigerado

    def mostrar_info(self):
        super().mostrar_info()
        estado = 'Sí' if self.refrigerado else 'No'
        print(f"  Tipo: Furgón | Refrigerado: {estado}")

    def calcular_costo(self, km):
        tarifa = 0.45 if self.refrigerado else 0.35
        return round(km * tarifa, 2)

    def tipo_carga(self):
        print(f"{self.placa}: Flores y productos perecibles.")


class Camion(Transporte):
    def __init__(self, placa, capacidad_kg, ejes):
        super().__init__(placa, capacidad_kg)
        self.ejes = ejes

    def mostrar_info(self):
        super().mostrar_info()
        print(f"  Tipo: Camión | Ejes: {self.ejes}")

    def calcular_costo(self, km):
        return round(km * 0.60, 2)

    def tipo_carga(self):
        print(f"{self.placa}: Carga pesada y maquinaria.")


class Motocicleta(Transporte):
    def __init__(self, placa, capacidad_kg, zona_entrega):
        super().__init__(placa, capacidad_kg)
        self.zona_entrega = zona_entrega

    def mostrar_info(self):
        super().mostrar_info()
        print(f"  Tipo: Moto | Zona: {self.zona_entrega}")

    def calcular_costo(self, km):
        return round(km * 0.18, 2)

    def tipo_carga(self):
        print(f"{self.placa}: Documentos y paquetes pequeños.")


# Lista de transportes - polimorfismo
flota = [
    Furgon("PBK-1234", 2000, True),
    Furgon("GHT-5678", 1500, False),
    Camion("TRE-9012", 8000, 6),
    Motocicleta("MCA-3456", 50, "Norte de Quito"),
]

km_ruta = 120

print("===== Flota Logística Bellaflor Group =====")
print(f"Ruta estimada: {km_ruta} km")
print("-" * 48)
for vehiculo in flota:
    vehiculo.mostrar_info()
    vehiculo.tipo_carga()
    print(f"  Costo ruta : ${vehiculo.calcular_costo(km_ruta):.2f}")
    print("-" * 48)
```
